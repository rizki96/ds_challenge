```python
import pandas as pd
```

```python
trf_df = pd.read_csv('training.csv')
```

```python
trf_df
```

```python
trf_df_sort = trf_df.sort_values(by=['day', 'timestamp'])
```

```python
trf_df_sort
```

```python
import geohash
import datetime
from sklearn import preprocessing
```

```python
trf_df_sort["daytimestamp_str"] = [str(datetime.timedelta(days=day, hours=int(timestamp.split(":")[0]), minutes=int(timestamp.split(":")[1]))) for day,timestamp in trf_df_sort[["day", "timestamp"]].values]
trf_df_sort["daytimestamp"] = [datetime.timedelta(days=day, hours=int(timestamp.split(":")[0]), minutes=int(timestamp.split(":")[1])) for day,timestamp in trf_df_sort[["day", "timestamp"]].values]
trf_df_sort["dts_cal"] = [datetime.datetime.strptime('2019-01-31 00:00:00.00', '%Y-%m-%d %H:%M:%S.%f') + datetime.timedelta(days=day, hours=int(timestamp.split(":")[0]), minutes=int(timestamp.split(":")[1])) for day,timestamp in trf_df_sort[["day", "timestamp"]].values]
```

# make sure demand value is normalized [0,1]

```python
scaler = preprocessing.MinMaxScaler()
trf_df_sort['mmnorm_demand'] = scaler.fit_transform(trf_df_sort[['demand']])
```

```python
dt_count = {dtstamp[0].split(',')[0]: 0 for dtstamp in trf_df_sort[['daytimestamp_str']].values}
```

```python
for dtstamp in trf_df_sort[['daytimestamp_str']].values:
    dt_count[dtstamp[0].split(',')[0]] += 1
```

```python
dt_count
```

# time granularity column in days

```python
trf_df_sort['dts_daily'] = [dtstamp[0].split(',')[0] for dtstamp in trf_df_sort[['daytimestamp_str']].values]
```

```python
trf_df_sort.loc[:, 'latlon'] = trf_df_sort.geohash6.map(lambda x: geohash.decode(x))
trf_df_sort['lat'] = [ll[0][0] for ll in trf_df_sort[['latlon']].values]
trf_df_sort['lon'] = [ll[0][1] for ll in trf_df_sort[['latlon']].values]
```

# day 1 cluster

```python
trf_day1 = pd.DataFrame(trf_df_sort[trf_df_sort.dts_daily == '1 day'])
```

```python
trf_day1.sort_values(by=['daytimestamp', 'demand'])
```

# testing for day 1 cluster

```python
import numpy as np
from sklearn.cluster import DBSCAN
coords = trf_day1[['lat', 'lon']].values
```

```python
kms_per_radian = 6371.0088
epsilon = 1 / kms_per_radian
db = DBSCAN(eps=epsilon, min_samples=1, algorithm='ball_tree', metric='haversine').fit(np.radians(coords))
cluster_labels = db.labels_
num_clusters = len(set(cluster_labels))
clusters = pd.Series([coords[cluster_labels == n] for n in range(num_clusters)])
print('Number of day 1 clusters: {}'.format(num_clusters))
```

```python
coords[cluster_labels == 109]
```

```python
from shapely.geometry import MultiPoint
from geopy.distance import great_circle
```

```python
def get_centermost_point(cluster):
    centroid = (MultiPoint(cluster).centroid.x, MultiPoint(cluster).centroid.y)
    centermost_point = min(cluster, key=lambda point: great_circle(point, centroid).m)
    return tuple(centermost_point)
centermost_points = clusters.map(get_centermost_point)
centermost_points
```

```python
lats, lons = zip(*centermost_points)
rep_points = pd.DataFrame({'lon':lons, 'lat':lats})
```

```python
rs = rep_points.apply(lambda row: trf_day1[(trf_day1['lat']==row['lat']) & (trf_day1['lon']==row['lon'])].iloc[0], axis=1)
rs
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=[10, 6])
rs_scatter = ax.scatter(rs['lon'], rs['lat'], c='#99cc99', edgecolor='None', alpha=0.7, s=120)
df_scatter = ax.scatter(trf_day1['lon'], trf_day1['lat'], c='k', alpha=0.9, s=3)
ax.set_title('Full data set vs DBSCAN reduced set')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.legend([df_scatter, rs_scatter], ['Full set', 'Reduced set'], loc='lower right')
plt.show()
```

# connection between cluster and demand on day 1 cluster

```python
trf_day1['cluster_ids'] = cluster_labels
```

```python
trf_day1['is_centroid'] = [(rs['geohash6'][y] == x) for x,y in trf_day1[['geohash6', 'cluster_ids']].values]
trf_day1 = trf_day1.sort_values(by=['cluster_ids', 'daytimestamp', 'demand'])
trf_day1
#trf_day1[(trf_day1.cluster_ids == 1) & (trf_day1.geohash6 == 'qp0dhu')]
#trf_day1[trf_day1.is_centroid == True]
#trf_day1[trf_day1.highest_demand == True]
#set([x for x,y in trf_day1[['cluster_ids', 'is_centroid']].values if y == True])
#trf_day1[trf_day1.cluster_ids == 0]
```

```python
trf_day1_cl0 = pd.DataFrame(trf_day1[trf_day1.cluster_ids == 1])
```

# centermost_point for day 1 cluster

```python
trf_day1_cl0[trf_day1_cl0.is_centroid == True]
```

```python
import seaborn as sns
sns.set(style="darkgrid")
sns.set(rc={'figure.figsize':(15.7,8.27)})
```

```python
trf_day1_cl0_center = pd.DataFrame(trf_day1_cl0[trf_day1_cl0.is_centroid == True])
trf_day1_cl0_others = pd.DataFrame(trf_day1_cl0[(trf_day1_cl0.is_centroid == False) & (trf_day1_cl0.demand > 0.0)])
#trf_day1_cl0_center = trf_day1_cl0_center[['dts_cal', 'geohash6', 'demand']].set_index(trf_day1_cl0_center.dts_cal)
#trf_day1_cl0_res_hourly = trf_day1_cl0_center.resample('H', on='dts_cal').mean()
#trf_day1_cl0_res_hourly.plot(style=[':', '--', '-'])
#data_set = sns.load_dataset(trf_day1_cl0_res_hourly)
ax = sns.lineplot(x="dts_cal", y="demand", hue="geohash6", data=trf_day1_cl0_others, legend="full")
for l in ax.lines:
    l.set_linestyle("--")
    plt.setp(l,linewidth=1)
sns.lineplot(x="dts_cal", y="demand", hue="geohash6", data=trf_day1_cl0_center, legend="full")
plt.show()
```

# highestdemand_point for day 1 cluster

```python
trf_day1_cl0_avg_demand = trf_day1_cl0.groupby('geohash6')['demand'].apply(lambda x: np.mean(x)).sort_values(ascending=False)
trf_day1_cl0['highest_demand'] = [trf_day1_cl0_avg_demand.index[0] == x[0] for x in trf_day1_cl0[['geohash6']].values]
trf_day1_cl0['avg_demand'] = [trf_day1_cl0_avg_demand[x[0]] for x in trf_day1_cl0[['geohash6']].values]
```

```python
trf_day1_cl0[trf_day1_cl0.highest_demand == True]
```

```python
trf_day1_cl0_highest = pd.DataFrame(trf_day1_cl0[trf_day1_cl0.highest_demand == True])
trf_day1_cl0_lower = pd.DataFrame(trf_day1_cl0[(trf_day1_cl0.highest_demand == False) & (trf_day1_cl0.demand > 0.0)])
ax = sns.lineplot(x="dts_cal", y="demand", hue="geohash6", data=trf_day1_cl0_lower, legend="full")
for l in ax.lines:
    l.set_linestyle("--")
    plt.setp(l,linewidth=1)
sns.lineplot(x="dts_cal", y="demand", hue="geohash6", data=trf_day1_cl0_highest, legend="full")
plt.show()
```
